# Temperature Prediction

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('../Datasets/SzegedWeatherHistory.csv')
```

```python
dataset.info()
```

```python
dataset.head(10)
```

```python
dataset['Formatted Date'] = pd.to_datetime(dataset['Formatted Date'],utc=True)
```

```python
dataset.head()
```

```python
dataset['year'] = dataset['Formatted Date'].dt.year
dataset['month'] = dataset['Formatted Date'].dt.month
dataset['day'] = dataset['Formatted Date'].dt.day
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
del dataset['Loud Cover']
```

```python
X = dataset.drop(['Daily Summary', 'Formatted Date'], axis=1)
y = dataset['Daily Summary']
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing  import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
categorical_indices = [0, 1]
```

```python
ct = ColumnTransformer(
    transformers = [('encoder',OneHotEncoder(),categorical_indices)],
    remainder = 'passthrough'
)
```

```python
X_transform = np.array(ct.fit_transform(X))
```

```python
print(X_transform)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_transform,y,test_size=0.2,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train,y_train)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN_classifier = KNeighborsClassifier(n_neighbors = 5)
```

```python
KNN_classifier.fit(X_train,y_train)
```

```python
y_pred = KNN_classifier.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```
