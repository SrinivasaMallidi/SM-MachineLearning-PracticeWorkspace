```python
#importing the required labaries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading the dataset 

dataset = pd.read_csv('../Datasets/SzegedWeatherHistory.csv')
```

```python
#displaying the dataset first 10 records 

dataset.head(10)
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
# Need to process the dataset. 

# Loud Cover coloumn is 0 for all the samples and hence can be ignored for training and testing.

# Formatted Date to be converted to three seperate coloumns Year, month, date 

# Encoding the catagorical date i.e  Summary , Precip Type

del dataset['Loud Cover']
```

```python
dataset.describe()
```

```python
# Formatted Date to be converted to three seperate coloumns Year, month, date

dataset['Formatted Date'] = pd.to_datetime(dataset['Formatted Date'],utc=True)
```

```python
dataset['year'] = dataset['Formatted Date'].dt.year
dataset['month'] = dataset['Formatted Date'].dt.month
dataset['day'] = dataset['Formatted Date'].dt.day
```

```python
dataset.head()
```

```python
# chossing the X and y

X = dataset.drop(['Formatted Date', 'Temperature (C)'], axis=1)
y = dataset['Temperature (C)']
```

```python
X
```

```python
y
```

```python
#  Encoding the catagorical date i.e  Summary , Precip Type

from sklearn.preprocessing import LabelEncoder
```

```python
# Select categorical columns
categorical_columns = ['Precip Type', 'Summary']

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Apply LabelEncoder to categorical columns
for column in categorical_columns:
    X[column] = label_encoder.fit_transform(X[column])
```

```python
print(X)
```

```python
X.describe()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
```

```python
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import  r2_score

accuracy = r2_score(y_test,y_pred)
```

```python
print('Accuracy {:.2f}{}'.format(accuracy*100,'%'))
```
