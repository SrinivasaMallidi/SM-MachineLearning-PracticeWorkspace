```python
import matplotlib.pyplot as plt
import numpy as np

# Define the grid size and center
grid_size = 9
center = (4,4)

# Create a grid with zeros
grid = np.zeros((grid_size, grid_size))

# Define the Vastu Purusha Mandala grid points
mandala_points = [(0,0), (2,2), (6,2), (8,0), (4,4), (8,8), (6,6), (2,6), (0,8)]

# Define the elements and symbols for the Mandala
symbols = [(1,4,'A'), (4,1,'B'), (4,7,'C'), (7,4,'D')]
boundary = [(3,3), (3,5), (5,5), (5,3), (3,3)]
entrance = [(0,3), (1,3), (1,5), (0,5)]

# Place the Vastu Purusha Mandala grid points on the main grid
for point in mandala_points:
    x, y = point
    grid[x,y] = 1

# Place the symbols on the Mandala
for symbol in symbols:
    x, y, label = symbol
    grid[x,y] = 1
    plt.text(y, x, label, fontsize=20, ha='center', va='center')

# Place the boundary on the Mandala
boundary_x = [point[0] for point in boundary]
boundary_y = [point[1] for point in boundary]
plt.plot(boundary_y, boundary_x, color='black', lw=2)

# Place the entrance on the Mandala
entrance_x = [point[0] for point in entrance]
entrance_y = [point[1] for point in entrance]
plt.plot(entrance_y, entrance_x, color='red', lw=2)

# Plot the grid
plt.imshow(grid, cmap='gray')
plt.xticks(np.arange(grid_size))
plt.yticks(np.arange(grid_size))
plt.grid(color='black', lw=2)
plt.title('Vastu Purusha Mandala')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the radius of the protractor
radius = 25

# Create a figure and axis object
fig, ax = plt.subplots(figsize=(6,6))

# Set the x and y limits
ax.set_xlim(-radius-1, radius+1)
ax.set_ylim(-radius-1, radius+1)

# Draw the circle
circle = plt.Circle((0,0), radius, fill=False, color='black')
ax.add_artist(circle)

# Draw the radius lines
for angle in range(0, 360, 5):
    x = radius * np.cos(np.deg2rad(angle))
    y = radius * np.sin(np.deg2rad(angle))
    ax.plot([0,x], [0,y], color='black', lw=0.5)
    
# Draw the degree labels
for angle in range(0, 360, 5):
    x = (radius + 0.5) * np.cos(np.deg2rad(angle))
    y = (radius + 0.5) * np.sin(np.deg2rad(angle))
    ax.text(x, y, str(angle), fontsize=10, ha='center', va='center')

# Draw the center point
ax.plot(0, 0, marker='o', markersize=5, color='black')

# Set the aspect ratio to be equal
ax.set_aspect('equal')

# Add the title and show the plot
ax.set_title('360-Degree Protractor')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# create a circle with radius 1 and center (0, 0)
circle = plt.Circle ((0, 0), 1, fill=False)

# create an arc with angle 360 degrees and center (0, 0)
arc = plt.Arc((0, 0), 2, 2, angle=0, theta1=0, theta2=360)

# create an array of angles from 0 to 360 degrees
angles = np.linspace (0, 360, 37)

# create an array of x and y coordinates for the angle marks
x = np.cos (np.radians (angles))
y = np.sin (np.radians (angles))

# plot the circle, arc and angle marks
fig, ax = plt.subplots ()
ax.add_patch (circle)
ax.add_patch (arc)
ax.plot (x, y, 'o')

# set the axis limits and aspect ratio
ax.set_xlim (-1.5, 1.5)
ax.set_ylim (-1.5, 1.5)
ax.set_aspect ('equal')

# add some labels for the angles
for i in range (len (angles)):
    ax.text (x [i], y [i], f'{angles [i]:.0f}', ha='center', va='center')

plt.show ()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches

# create a circle with radius 1 and center (0, 0)
circle = plt.Circle ((0, 0), 1, fill=False)

# create an arc with angle 360 degrees and center (0, 0)
arc = patches.Arc ((0, 0), 2, 2, angle=0, theta1=0, theta2=360)

# create an array of angles from 360 to 0 degrees and add 90 degrees
angles = np.linspace (360, 0, 37) + 90

# create an array of x and y coordinates for the angle marks
x = np.cos (np.radians (angles))
y = np.sin (np.radians (angles))

# plot the circle, arc and angle marks
fig, ax = plt.subplots (figsize=(25, 25))
ax.add_patch (circle)
ax.add_patch (arc)
ax.plot (x, y, 'o')

# set the axis limits and aspect ratio
ax.set_xlim (-1.5, 1.5)
ax.set_ylim (-1.5, 1.5)
ax.set_aspect ('equal')

# add some labels for the angles
for i in range (len (angles)):
    # subtract the angle from 360 and add 90 to get the clockwise angle
    ax.text (x [i], y [i], f'{360 - angles [i] + 90:.0f}', ha='center', va='center')

plt.show ()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches

# create a circle with radius 1 and center (0, 0)
circle = plt.Circle ((0, 0), 10, fill=False)
circle1 = plt.Circle ((0, 0), 1, fill=False)

# create an arc with angle 360 degrees and center (0, 0)
arc = patches.Arc ((0, 0), 2, 2, angle=0, theta1=0, theta2=360)

# create an array of angles from 360 to 0 degrees and add 90 degrees
angles = np.linspace (360, 0, 37) + 90

# create an array of x and y coordinates for the angle marks
x = np.cos (np.radians (angles))
y = np.sin (np.radians (angles))

# plot the circle, arc and angle marks
fig, ax = plt.subplots (figsize=(25, 25))
ax.add_patch (circle)
ax.add_patch (circle1)
ax.add_patch (arc)
ax.plot (x, y, 'o')

# set the axis limits and aspect ratio
ax.set_xlim (-1.5, 1.5)
ax.set_ylim (-1.5, 1.5)
ax.set_aspect ('equal')

# add some labels for the angles
for i in range (len (angles)):
    # subtract the angle from 360 and add 90 to get the clockwise angle
    ax.text (x [i], y [i], f'{360 - angles [i] + 90:.0f}', ha='center', va='center')

plt.show ()
```
