## Accessing Data

Extracting data present in defined data structures is known as Data Access. Pandas offers utilities such as loc and iloc for retrieving data from a Series, DataFrame, or Panel.

To access individual elements, you can use square brackets and specify either the index number or index value.

```python
np_array = np.arange(0,10,2)
pd_series = pd.Series(np_array,index=list('abcde'))

print(pd_series)
```

```python
#Accessing 3rd element of pd_series. 
print(pd_series[2]) 
#Accessing 4th element of pd_series. 
print(pd_series['d']) 
```

Using the get method, it is possible to access a single element by passing the index number or index value as an argument.

```python
print(pd_series.get(1))
print(pd_series.get('e'))
```

You can slice a Pandas Series in a manner that is quite similar to how you would slice a Python list. When using index values to slice a Pandas Series, the elements corresponding to both the start and end index values are included in the resulting slice.

```python
print(pd_series[1:4])
```

```python
print(pd_series['a':'c'])
```

In Pandas, you can use the describe() method to get summary statistics for a Series object. The describe() method returns a DataFrame that includes count, mean, standard deviation, minimum, maximum, and quartile values for the data in the Series.

Here is an example of how to use describe() method:

```python
# use the describe() method to get summary statistics
series_stats = pd_series.describe()
print(series_stats)
```

you can use the info() and describe() methods to get an overview of the data present in a DataFrame object.

The info() method provides a summary of the DataFrame, including the column names, the number of non-null values in each column, and the data types of each column. Here is an example of how to use the info() method:

```python
data = {'name': ['Srinivasa', 'Priya', 'Arvind', 'Swapna'], 
        'age': [25, 30, 28, 21], 
        'gender': ['M', 'F', 'M', 'F']}
df = pd.DataFrame(data)

# use the info() method to get a summary of the DataFrame
df.info()
```

As you can see, the info() method provides useful information about the DataFrame, including the number of entries, the column names, the number of non-null values in each column, and the data types of each column.

The describe() method provides a statistical summary of the DataFrame, including the count, mean, standard deviation, minimum, and maximum values, as well as the quartile values for each numeric column in the DataFrame. Here is an example of how to use the describe() method:

```python
# use the describe() method to get a statistical summary of the DataFrame
df.describe()
```

As you can see, the describe() method provides useful statistical information about the DataFrame, including the count, mean, standard deviation, minimum, and maximum values, as well as the quartile values for the numeric column in the DataFrame.

## Crosstab 

Crosstab  is a function in Pandas that computes a cross-tabulation of two (or more) categorical variables. It takes one or more arrays, DataFrame, or Series as input, and returns a DataFrame that shows the frequency distribution of the variables in tabular form. The resulting table provides a breakdown of how many times each combination of variables occurs in the data.

```python
# using crosstab to get frequency distribution of driver_gender and violation
freq_table = pd.crosstab(trafficdata_output['driver_gender'], trafficdata_output['violation'])

print(freq_table)
```

```python
# using groupby to get mean driver_age by driver_gender
age_by_gender = trafficdata_output.groupby('driver_gender')['driver_age'].mean()

print(age_by_gender)
```

## .fillna()

This function is used to fill missing or NaN values in a DataFrame.

Syntax: DataFrame.fillna(value=None, method=None, axis=None, inplace=False, limit=None, downcast=None)

    * value: scalar, dict, Series, or DataFrame - The value to use for missing values.

    * method: string, optional (default None) - Method to use for filling holes in reindexed Series pad / ffill: propagate last valid observation forward to next valid backfill / bfill: use next valid observation to fill gap

    * axis: {0 or 'index', 1 or 'columns'}, default None - Axis along which to fill missing values.

    * inplace: bool, default False - Modify the DataFrame in place (do not create a new object).

    * limit: int, optional - If method is specified, this is the maximum number of consecutive NaN values to forward/backward fill. In other words, if there is a gap with more than this number of consecutive NaNs, it will only be partially filled.

    * downcast: dict, optional - A dict of item->dtype of what to downcast if possible, or the string 'infer' which will try to downcast to an appropriate equal type (e.g. float64 to int64 if possible).

```python
trafficdata_output['county_name']
```

```python
trafficdata_output['county_name'].fillna(0) # not using inplace
```

```python
trafficdata_output['county_name'] 
```

## .dropna()

 This function is used to remove rows or columns with missing values from a DataFrame.

Syntax: DataFrame.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)

    * axis: int or str, optional (default=0) - Axis to use.

    * how: {any, all}, optional (default=any) - Determine if row or column is removed from DataFrame, when we have at least one NA or all NA.

    * thresh: int, optional - Require that many non-NA values.

    * subset: array-like, optional - Labels along other axis to consider, e.g. if you are dropping rows these would be a list of columns to include.

    * inplace: bool, default False - Modify the DataFrame in place (do not create a new object).

# .apply()

The .apply() method in pandas is used to apply a function to a dataframe or a series of a dataframe. The function can be a built-in function or a custom function.

For example, let's say we want to create a new column in the trafficdata_output dataframe called driver_age_group which categorizes drivers based on their age. We can define a function that takes in an age value and returns a string indicating the age group

```python
def categorize_age(age):
    if age < 25:
        return 'Under 25'
    elif age >= 25 and age < 40:
        return '25-39'
    elif age >= 40 and age < 60:
        return '40-59'
    else:
        return '60 and over'
```

Then we can use the .apply() method to apply this function to the driver_age column of the trafficdata_output dataframe and create a new column called driver_age_group

```python
trafficdata_output['driver_age_group'] = trafficdata_output['driver_age'].apply(categorize_age)
```

```python
trafficdata_output
```

Note: that the apply method can also be used to apply a function row-wise or column-wise by setting the axis parameter to 0 or 1, respectively. In this case, the function would be applied to each row or column of the dataframe.

If we want to calculate the mean of each column, we can use the apply() function with axis=0 parameter as follows:

```python
# create a dataframe
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})

# define a function to calculate the sum of a row or column
def sum_func(x):
    return sum(x)

print(df)
```

```python
# apply the sum_func row-wise
row_sum = df.apply(sum_func, axis=1)
print(row_sum)
```

```python
# apply the sum_func column-wise
col_sum = df.apply(sum_func, axis=0)
print(col_sum)
```
